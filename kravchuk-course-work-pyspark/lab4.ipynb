```
import pyspark.sql.functions as F
from pyspark.sql.window import *
```

Створюємо схему та зчитуємо дані з файлу

```
schema = "`ID` INT,`Length` INT,`Birthweight` DOUBLE,`Headcirc` INT,`Gestation` INT,`smoker` INT,`mage` INT,`mnocig` INT,`mheight` INT,`mppwt` INT,`fage` INT,`fedyrs` INT,`fnocig` INT,`fheight` INT,`lowbwt` INT,`mage35` INT"

df = spark.read.format('csv').option('header', True).option('schema', schema).load("/FileStore/tables/Birthweight__1_.csv")
display(df)
```

1) Створити нову таблицю, яка буде містити наступні два стовпці: вік
матері; масив ІD записів з дітьми матерів відповідного віку.

```
df1 = df.groupBy("mage") \
        .agg(F.collect_list("ID") \
        .alias("ID")) \
        .withColumnRenamed("mage", "Вік") 

display(df1)
```

2) Взяти таблицю з завдання 1 і перетворити другий стовпець на кілька
стовпців.

```
df2 =df1 \
.select('Вік', \
        F.explode('ID').alias('id'), \
        F.row_number().over(Window.partitionBy("Вік").orderBy("Вік")).alias('number'))

display(df2)
```

```
df2 = df2 \
    .groupby('Вік') \
    .pivot('number') \
    .agg(F.first('id'))

display(df2)
```

3) Знайти середню вагу дітей для усіх можливих поєднань параметрів
smoker та mage35.

```
df3 = df \
    .groupBy("smoker", "mage35") \
    .agg(F.avg("Birthweight").alias('AverageWeight'))
display(df3)
```

4) Додати в таблицю Birthweight новий стовпець, що містить різницю
між середнім та поточним значенням ваги дитини в залежності від
того чи палить батько. Наприклад, якщо середня вага дітей, у яких
батько не палить, 3.55, а в першому рядку вага 3.5 і батько не палить,
то значення в додатковому стовпчику = 3.55-3.5=0.05. Зберегти
новий файл.

```
df4 = df.withColumn("fsmoker", F.when((F.col("fnocig") > 0), 1).otherwise(0))
display(df4)
```

```
df4_tmp = df4 \
    .groupby("fsmoker") \
    .agg(F.avg("BirthWeight").alias('AverageWeight')) \
    .withColumnRenamed("fsmoker", "indicator")

display(df4_tmp)
```

```
df4 = df4 \
    .join(df4_tmp, df4.fsmoker ==  df4_tmp.indicator, "inner") \
    .withColumn("result",F.col("AverageWeight") - F.col("Birthweight")) \
    .select("ID", "Birthweight", "AverageWeight", "fsmoker", "result")
display(df4)
```

```
df4.write.csv("/FileStore/tables/result.csv")
```
