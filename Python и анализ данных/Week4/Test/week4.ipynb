Далее оценим распределение выборочного среднего при разных объёмах выборок(n = 5, 10, 15).
Для гамма-распределения с параметром 10 (Г(10)) математическое ожидание и дисперсия равны этому параметру, то есть 10.
Значит, по центральной предельной теореме выборочное среднее стремится к нормальному с параметрами $N(10,\frac{10}{n})$

```python
n = 5

#функция, реализующая выборочное среднее
def s_mean(x):
    s = 0.
    for i in range(len(x)):
        s += x[i]
    return s/len(x)

#расчёт выборочных средних для построения гистограммы
s_mean_list = [0 for i in range(1000)]
for i in range(1000):
    s_mean_list[i] = s_mean(gamma_rv.rvs(n))

#нормальное распределение, к которому стремится выборочное среднее, и построение графика плотности данного распределения    
norm_rv = sts.norm(10, 10./n)
x = np.linspace(5,17,1500)
norm_pdf = norm_rv.pdf(x)
plt.plot(x, norm_pdf, 'r-', label='theoretical PDF')
    
#построение гистограммы
plt.hist(s_mean_list, ec="black", normed=True)

plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend();
```
